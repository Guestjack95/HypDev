Compare the average income based on ethnicity:

```python
ethnicity = df['Ethnicity'].unique()
for value in ethnicity:
    print(value)
```

```python
print("The average income of Caucasians is", df[df.Ethnicity == "Caucasian"].loc[:,"Income"].mean())
print("The average income of Asians is", df[df.Ethnicity == "Asian"].loc[:,"Income"].mean())
print("The average income of African Americans is", df[df.Ethnicity == "African American"].loc[:,"Income"].mean())
```

On average, do married or single people have a higher balance? 

```python
df.head()
```

```python
print("The average balance of Married people is", df[df.Married == "Yes"].loc[:,"Balance"].mean())
print("The average balance of Unmarried people is", df[df.Married == "No"].loc[:,"Balance"].mean())
```

What is the highest income in our dataset?

```python
df.sort_values(by='Income',ascending=False).head(1)
```

```python
df.loc[:, 'Income'].max()
```

What is the lowest income in our dataset?

```python
df.sort_values(by='Income',ascending=True).head(1)
```

```python
df.loc[:, 'Income'].min()
```

How many cards do we have recorded in our dataset?

```python
df['Cards'].sum()
```

How many females in our dataset?
How many males in our dataset?

```python
print("Females\n",df[df.Gender == 'Female'].count())
print("Males\n",df[df.Gender == 'Male'].count())
```

```python
female = df['Gender'][df['Gender'] == 'Female'].count()
male = df['Gender'][df['Gender'] == 'Male'].count()

print(f"The number of females in this dataset is {female}")
print(f"The number of males in this dataset is {male}")
```
